```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
y = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, y)
Z = np.random.random((100, 100))

plt.contourf(X, Y, Z, cmap='Pastel1')
plt.colorbar()
plt.title('Homework1 -- ContourPlot')
plt.xlabel('X')
plt.ylabel('Y')

plt.show()

plt.savefig('HW1ContourPlot.png')
```
